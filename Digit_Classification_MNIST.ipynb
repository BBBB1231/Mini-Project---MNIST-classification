```python
# Importing libraries
import tensorflow as tf
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

Loading the data

```python
(X_train, Y_train), (X_test, Y_test) = tf.keras.datasets.mnist.load_data() #The data we need in our model
```

Visualizing the variables: Looking at X_train and Y_train

```python
plt.imshow(X_train[0], cmap='gray') #The gray will detect how 'gray' a block is comparing to one another to get the shape of the image
Y_train[0]
```

Building the model, the fun(but complicated) part

```python
#Declaring the type of model
model = tf.keras.models.Sequential()

#Adding Layers to the model
model.add( layers.Input(np.shape(X_train)[1:]) )
model.add( layers.Flatten() )
model.add( layers.Dense(16, activation = 'relu') ) #What is relu?
model.add( layers.Dense(16, activation = 'relu') )
#Final Layer, where the prediction will be made
model.add( layers.Dense(10, activation = 'softmax') )

#Building the model
model.compile(optimizer='adam', metrics = ['accuracy'], loss = 'sparse_categorical_crossentropy')
```

```python
#Training the model
model.fit(X_train, Y_train, epochs = 10, batch_size = 32) #epochs: time period
```

How good is our model?

```python
model.evaluate(X_test, Y_test)
```

View Some Predictions

```python
number_of_images = 5

probabilities = model.predict(X_train[:number_of_images])
predictions = np.argmax(probabilities, axis=1)
for index in range(len(probabilities)):
  plt.imshow(X_train[index], cmap='gray')
  plt.show()
  print("Prediction: ", predictions[index])
```

What is it getting wrong?

```python
total_predictions = model.predict(X_test)
for i in range(len(total_predictions)):
  prediction = np.argmax(total_predictions[i])
  label = Y_test[i]
  if prediction != label:
    plt.imshow(X_test[i], cmap='gray')
    plt.show()
    print("Prediction: ", prediction)
  if i > 1000:
    break
```

```python
#shaping the data

X_train = X_train / 255
X_train.reshape(60000, 28, 28, 1)
X_test = X_test / 255
np.shape(X_test)
```

Model Improvement

```python
#building a new model using convolutions

goated_model = tf.keras.models.Sequential()

goated_model.add( layers.Conv2D(64, (3, 3), activation = 'relu', input_shape = (28, 28, 1)) )
goated_model.add( layers.MaxPool2D((2, 2)))

goated_model.add( layers.Conv2D(64, (3, 3), activation = 'relu') )
goated_model.add( layers.MaxPool2D((2, 2)))

goated_model.add( layers.Flatten() )
goated_model.add( layers.Dense(128, activation = 'relu'))
goated_model.add( layers.Dense(10, activation = 'softmax'))

goated_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
goated_model.fit(X_train, Y_train, epochs = 3, batch_size = 32)
```

```python
goated_model.evaluate(X_test, Y_test)
```
